Grocery Stock Tracker:
available_stock = ["apple","milk","bread","eggs","rice","butter"]
Tasks:
Add "flour" and "sugar" at the end.
Replace "bread" with "brown bread".
Remove "milk" and print the updated list.
Display only the middle 3 items using slicing.

Grocery Stock Tracker:
ticket_prices = [250, 300, 200, 400, 150]
Tasks:
Find the highest and lowest ticket price.
Sort the list in descending order.
Add a new ticket price 350 while keeping the list sorted.
Display only the top 3 expensive tickets.

```python
ticket_prices = [250, 300, 200, 400, 150]

highest = max(ticket_prices)
lowest = min(ticket_prices)

print("Highest price:", highest)
print("Lowest price:", lowest)

ticket_prices.sort(reverse=True)
print("Descending order:", ticket_prices)

ticket_prices.append(350)
ticket_prices.sort(reverse=True)
print("After adding 350:", ticket_prices)

top_3 = ticket_prices[:3]
print("Top 3 expensive tickets:", top_3)
```

Product Prices:
You have prices of 6 products in your shop.
Create a list of product prices.
Increase the price of two products due to tax (1% tax /product).
Remove the lowest-priced product.
Calculate the total price of the remaining products.

```python
prices = [120, 250, 180, 300, 90, 200]

prices[0] = prices[0] + prices[0] * 0.01
prices[1] = prices[1] + prices[1] * 0.01

lowest = min(prices)
prices.remove(lowest)

total = sum(prices)

print("Remaining prices:", prices)
print("Total price:", total)
```

Apartment Details:
An apartment listing is stored as a tuple:
apartment = ("Downtown"
, 3, 2, 45000,
"Available")
- Extract and print the location and rent.
- Change the status to "Rented" without modifying the original tuple.

```python
apartment = ("Downtown", 3, 2, 45000, "Available")

location = apartment[0]
rent = apartment[3]

print(f"Location: {location}")
print(f"Monthly Rent: ${rent}")

updated_apartment = apartment[:4] + ("Rented",)

print(f"Original: {apartment}")
print(f"Updated:  {updated_apartment}")
```

Employee Payroll Record:
Payroll system keeps a tuple: (employee_id, name, base_salary, month, net_pay).
Design data for multiple employees and print the one with the highest salary for a given
month.
Note: Atleast 3 employee records must be created

```python
payroll_records = [
    (101, "Ayesha", 75000, "March", 62000),
    (102, "Saba", 82000, "March", 68000),
    (103, "Ali", 69000, "March", 57000)
]

target_month = "March"
highest_emp = None

for emp in payroll_records:
    if emp[3] == target_month:
        if highest_emp is None or emp[2] > highest_emp[2]:
            highest_emp = emp

if highest_emp:
    print(f"Highest Salary in {target_month}:")
    print(f"Name: {highest_emp[1]}")
    print(f"Salary: {highest_emp[2]}")
```
