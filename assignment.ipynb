Basic Salary Increment Calculator:
Ask the user:  
"Enter your current salary:"  
"Enter your bonus percentage:"  
Calculate bonus amount using percentage.
Update salary using += assignment operator.

```python
current_salary = float(input("Enter your current salary: "))
bonus_percentage = float(input("Enter your bonus percentage: "))
bonus_amount = current_salary * (bonus_percentage / 100)
current_salary += bonus_amount
print(f"Your bonus amount is: ${bonus_amount}")
print(f"Your updated salary is: ${current_salary}")
```

Gym Membership Fee Calculator
Ask the user:
Enter registration fee (integer).  
Enter monthly fee (integer).  
Enter number of months they want to pay for.
Calculate:
Total cost = registration fee plus (monthly fee × number of months).  
Then:
Ask if the user has a discount coupon (yes/no).  
If yes, reduce the total by 10% (use assignment operators, no if).
Print the final amount to be paid.
(Hint: Use *= for 10% reduction using 0.9 multiplier.)

```python
registration_fee = int(input("Enter registration fee: "))
monthly_fee = int(input("Enter monthly fee: "))
months = int(input("Enter number of months: "))

total_cost = registration_fee + (monthly_fee * months)
has_coupon = input("Do you have a discount coupon (yes/no)? ")
total_cost *= 0.9 if has_coupon == "yes" else 1
print(f"Final amount to be paid: ${total_cost}")
```

Simple E-Wallet Recharge and Spending:
Start with an e-wallet balance of 0.
Ask the user:
"How much money do you want to add?"  
"How much money did you spend on food?"  
"How much money did you spend on transport?"  
"How much money did you spend on shopping?"  
Use assignment operators (+=
, -=) to update the wallet after each step.
Then:
Calculate whether the balance is more than 500 and less than 5000 (use logical
operator to print True or False).
Print the final wallet balance.

```python
wallet = 0

wallet += float(input("How much money do you want to add? "))
wallet -= float(input("How much money did you spend on food? "))
wallet -= float(input("How much money did you spend on transport? "))
wallet -= float(input("How much money did you spend on shopping? "))

print(wallet > 500 and wallet < 5000)
print(f"Final wallet balance: ${wallet}")
```
